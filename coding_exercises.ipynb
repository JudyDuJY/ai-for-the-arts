# 5.Concept: data structures such as dictionaries
**Copilot initial prompt:**

**Follow-up prompts(if any)**

## My notes
A dictionary is a key-value collection enclosed in curly braces {}, where keys are usually strings and must be immutable. There are two ways to access it: one is [key] (which will throw an error if the key does not exist), and the other is .get(key, default value) (safe).
## The code I tested


```python

# 1) Define a dictionary that maps language names (keys) to a greeting (value)
greetings = {
    "English": "Hello, world!",
    "Chinese": "‰Ω†Â•ΩÔºå‰∏ñÁïåÔºÅ",
    "Spanish": "¬°Hola, mundo!",
    "French": "Bonjour, le monde!",
}

# 2) Pick a language we want to use
chosen_language = "Chinese"

# 3) Look up the greeting by key and print it
print(greetings[chosen_language])

# 4) Add a new key‚Äìvalue pair to the dictionary
greetings["German"] = "Hallo, Welt!"

# 5) Safely get a greeting for a language that may not exist (returns default if missing)
print(greetings.get("Japanese", "Sorry, that language is not available."))

# 6) Update an existing value
greetings["English"] = "Hello, Python world!"

# 7) Loop through the dictionary and print all greetings (key and value)
for language, message in greetings.items():
    print(f"{language}: {message}")
```

# 6.Concept:  data structures such as arrays
**Copilot initial prompt:**
Could you suggest a modification of the Python code for ‚ÄúHello, world!‚Äù to demonstrate the concept of arrays to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects.
**Follow-up prompts(if any)**
‚ÄúRemoving and inserting items‚Äù
## My notes
Arrays and lists differ. Arrays require all elements to be of the same type, which is more stringent. They also allow for chained numerical calculations, making them more efficient. Therefore, arrays are often used for numerical processing.
## The code I tested


```python

import array

# 1) Create an integer array (typecode 'i' means integer)
numbers = array.array('i', [1, 2, 3, 4, 5])
print(numbers)  # array('i', [1, 2, 3, 4, 5])

# 2) Insert an element
numbers.insert(2, 99)  # Insert 99 at index 2
print(numbers)  # array('i', [1, 2, 99, 3, 4, 5])

# 3) Remove an element by value
numbers.remove(3)  # Remove the element with value 3
print(numbers)  # array('i', [1, 2, 99, 4, 5])

# 4) Use pop to remove and return an element
removed = numbers.pop()  # Remove the last element
print("Removed element:", removed)
print(numbers)

# 5) Iterate through the array
for num in numbers:
    print(num)

# 6) Get the length of the array
print("Array length:", len(numbers))
```

# 9. Concept: how to use operators in Python
**Copilot initial prompt:**
Could you suggest a modification of the Python code for ‚ÄúHello, world!‚Äù to demonstrate the concept of operators to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects. Include discussions about addition and multiplication between strings and numbers.
## My notes
Different numeric operators allow you to combine different elements or repeat the same elements. Basic numeric operators include: +, -, *, / (floating-point division, keeping decimals), // (integer division, removing decimals), % (remainder), and ** (exponentiation). In string operations, + concatenates strings, and * repeats strings. The operator precedence is multiplication and division first, then addition and subtraction; parentheses take precedence.
## The code I tested


```python

# 9. Operators ‚Äî starting from "Hello, world!"

# 1) Start with the classic
print("Hello, world!")

# 2) Arithmetic operators with numbers
a = 10
b = 3
print("a + b =", a + b)    # addition
print("a - b =", a - b)    # subtraction
print("a * b =", a * b)    # multiplication
print("a / b =", a / b)    # true division (float)
print("a // b =", a // b)  # floor division (integer result)
print("a % b =", a % b)    # modulus (remainder)
print("a ** b =", a ** b)  # exponentiation

# 3) String operators: + and *
greeting = "Hello"
name = "world"
print(greeting + ", " + name + "!")  # string concatenation with +

echo = "ha"
print(echo * 3)  # string repetition: "hahaha"

# 4) Addition/multiplication between strings and numbers
num = 5
text = "5"

# Valid: number + number
print("num + 5 =", num + 5)

# ‚ùóInvalid: string + number (TypeError)
# print(text + 5)  # Uncommenting this line causes TypeError

# Valid: string + string
print('"5" + "5" =', text + text)

# Valid: string * number (repetition)
print('"hi" * 4 =', "hi" * 4)

# ‚ùóInvalid: number * string * string (only str * int is allowed; order (int * str) is fine too)
# print(3 * "hi" * "bye")  # TypeError

# 5) Comparison operators (return True/False)
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a <= b:", a <= b)

# 6) Logical operators with booleans
is_big = a > 5
is_small = b < 2
print("is_big and is_small:", is_big and is_small)
print("is_big or is_small:", is_big or is_small)
print("not is_small:", not is_small)

# 7) Assignment operators (shorthand updates)
x = 2
x += 3   # x = x + 3
x *= 4   # x = x * 4
print("x after += and *= is", x)

# 8) Operator precedence (PEMDAS-like; use parentheses to be clear)
print("3 + 2 * 5 =", 3 + 2 * 5)       # multiplication before addition -> 13
print("(3 + 2) * 5 =", (3 + 2) * 5)   # parentheses first -> 25
```

# 10. Concept: how to use advanced concepts such as classes, objects, and methods
**Copilot initial prompt:**
Could you suggest a modification of the Python code for ‚ÄúHello, world!‚Äù to demonstrate the concept of classes, objects and methods to a complete beginner in programming? Please explain the code step-by-step highlighting any key aspects.
## My notes
A class, like a blueprint or recipe, defines the structure and behavior of objects. An object (or instance) is a concrete thing created according to the blueprint. A method is a function within a class that represents an action that an object can perform. Within a method, ‚Äòself‚Äô refers to the current object. The difference between properties and methods is that properties are data (e.g., ‚Äòself.name‚Äô), while methods are behaviors (e.g., ‚Äògreet()‚Äô).
## The code I tested


```python

# 1) Define a class named Greeter
class Greeter:
    # 2) The constructor method: runs when you create a Greeter
    def __init__(self, name="world"):
        # 3) Store data in the object (an attribute)
        self.name = name

    # 4) A method: what the object can do
    def greet(self):
        print(f"Hello, {self.name}!")

    # 5) Another method to change the object's data
    def set_name(self, new_name):
        self.name = new_name

# -------- Using the class (creating objects) --------

# 6) Create a Greeter object (an instance of the class)
g = Greeter()           # uses the default name "world"
g.greet()               # Output: Hello, world!

# 7) Create another object with a custom name
g2 = Greeter("Alice")
g2.greet()              # Output: Hello, Alice!

# 8) Change the name of the first object and greet again
g.set_name("Python")
g.greet()               # Output: Hello, Python!
```
